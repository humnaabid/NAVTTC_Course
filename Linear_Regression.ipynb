# Changing the categorical data into Numerical through Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['mainroad'] = label.fit_transform(df['mainroad'])
df['guestroom'] = label.fit_transform(df['guestroom'])
df['basement'] = label.fit_transform(df['basement'])
df['hotwaterheating'] = label.fit_transform(df['hotwaterheating'])
df['airconditioning'] = label.fit_transform(df['airconditioning'])
df['prefarea'] = label.fit_transform(df['prefarea'])
df['furnishingstatus'] = label.fit_transform(df['furnishingstatus'])
```

# Removing of outlier from all dataset using a Function

```python
def mod_outlier(df):
    col_vals = df.columns
    df1 = df.copy()
    df = df._get_numeric_data()

    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)

    iqr = q3 - q1

    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    for col in col_vals:
        for i in range(0, len(df[col])):
            if df[col][i] < lower_bound[col]:
                df[col][i] = lower_bound[col]

            if df[col][i] > upper_bound[col]:
                df[col][i] = upper_bound[col]

    for col in col_vals:
        df1[col] = df[col]

        return(df1)
```

```python
mod_outlier(df)
```
